```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y=x**2
```

```python
plt.plot(x,y,'k--o')
plt.xlim(0,9)
plt.ylim(0,90)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Random plot")
plt.grid(True)
```

```python
mat = np.arange(0,100).reshape(10,10)
plt.imshow(mat,cmap='magma')
```

```python
mat = np.random.randint(0,1000,(10,10))
plt.imshow(mat)
plt.colorbar()
```
